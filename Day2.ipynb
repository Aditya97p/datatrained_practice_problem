```python
url = "https://github.com/training-ml/Files/raw/main/raw_data.csv"

df = pd.read_csv(url)
```

```python
new_data = {
    'id': 210,
    'Employee Full Name': 'Jim Carry',
    'DOB': '1/1/1970',
    'email': 'jim.carry@gmail.com',
    'gender': 'Male',
    'Company': 'Amazon',
    'Country': 'USA',
    'Income.1': 12349.556,
    'Income.2': 2365
}

# Use the loc method to insert the new row at index 9
df.loc[9] = new_data

# Print the DataFrame to verify the addition
print(df)




```

```python
new_emails = ['graeme.Croisier@gmail.com']
new_emails1 = ['Modesta.Batterson@gmail.com']


positions_to_add_emails = [3]

for i, position in enumerate(positions_to_add_emails):
    if pd.isna(df.loc[position, 'email']):
        df.loc[position, 'email'] = new_emails[i]

positions_to_add_emails = [6]
for i, position in enumerate(positions_to_add_emails):
    if pd.isna(df.loc[position, 'email']):
        df.loc[position, 'email'] = new_emails1[i]

# Print the updated DataFrame
print(df)
```
